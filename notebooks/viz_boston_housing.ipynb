```python
# Distribution plot

import warnings
warnings.filterwarnings("ignore")

rows = 2
cols = 7

fig, ax = plt.subplots(nrows=rows, ncols = cols, figsize = (20, 10))

col = data.columns
index = 0

for i in range(rows):
    for j in range(cols):
        sns.distplot(data[col[index]], ax = ax[i][j])
        index = index + 1
        
plt.tight_layout()
```

```python
# Scatter plot 

rows = 2
cols = 7

fig, ax = plt.subplots(rows, cols, figsize = (20, 10))

col = data.columns
index = 0

for i in range(rows):
    for j in range(cols):
        sns.scatterplot(x = 'Price', y = col[index], data = data, ax = ax[i][j])
        index = index + 1
        
plt.tight_layout()
plt.show()
```

```python
# Correlation matrix

corrmat = data.corr()
corrmat
```

```python
# Headmap plot

fig, ax = plt.subplots(figsize = (14, 10))
sns.heatmap(corrmat, annot = True, annot_kws = {'size': 7})

bottom, top = ax.get_ylim()
ax.set_ylim(bottom+0.5, top-.5)
plt.show()
```

```python
# Index correlation matrix

corrmat.index
```

```python
# Create corellated features

def getCorrelatedFeature(corrdata, threshold):
    feature = []
    value = []
    
    for i, index in enumerate(corrdata.index):
        if abs(corrdata[index]) > threshold:
            feature.append(index)
            value.append(corrdata[index])
            
    df = pd.DataFrame(data = value, index=feature, columns=['corr value'])
    
    return df
```

```python
# Treshold

threshold = 0.5
corr_df = getCorrelatedFeature(corrmat['Price'], threshold)
```

```python
# Correlation values

corr_df
```

```python
# Correlated data

correlated_data = data[corr_df.index]
correlated_data.head()
```

```python
# Pair plot 

sns.pairplot(correlated_data)
plt.tight_layout()
```

```python
# Correlation plot

fig, ax = plt.subplots(figsize = (4, 4))
sns.heatmap(correlated_data.corr(), annot = True, annot_kws = {'size': 12})

bottom, top = ax.get_ylim()
ax.set_ylim(bottom + 0.5, top - 0.5)
plt.show()
```
